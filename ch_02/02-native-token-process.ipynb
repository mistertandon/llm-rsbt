```python
import re
```

```python
text = "Hello, world. Is this-- a test?"
```

```python
result_ver_01 = re.split(r'(\s)', text)
```

```python
print(result_ver_01)
```

```python
result_ver_02 = re.split(r'([,.:;?_!"()\']|--|\s)', text)
print(result_ver_02)
```

```python
final_result = [ item for item in result_ver_02 if item.strip()]
print(final_result)
```
