No we'll try to compute attention score corresponding to q^2 i.e. journey

attention_score_2 is un-normalized attention score represneted by omega for 2nd element (At index 1)

```python
attention_score = torch.empty(inputs_embeddings.shape[0], inputs_embeddings.shape[0])
for i, inputs_embedding_i in enumerate(inputs_embeddings):
    for j, inputs_embedding_j in enumerate(inputs_embeddings):
        attention_score[i, j] = torch.dot(inputs_embedding_i, inputs_embedding_j)

print(attention_score)
```

We can obtain attention score using matrix multiplication as well

```python
attention_score_v2 = inputs_embeddings @ inputs_embeddings.T
print(attention_score_v2)
```

```python
attention_weight = torch.softmax(attention_score, dim=-1)
print(attention_weight)
```

```python
context_vector_all = attention_weight @ inputs_embeddings
print(context_vector_all)
```
